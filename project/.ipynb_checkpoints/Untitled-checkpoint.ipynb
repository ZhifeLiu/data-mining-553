```python
import sys
from pyspark.sql import SparkSession
from pyspark import SparkContext, SparkConf
import re
import unicodedata
from operator import add
from itertools import combinations
```

```python
spark = SparkSession \
            .builder \
            .appName("clustering") \
            .getOrCreate()
sc = spark.sparkContext
```

```python
data = spark.read.csv('inspections.csv', header = True)
```

```python
names = data.rdd.map(lambda x: x['facility_name']).distinct()\
                .map(lambda x: (x, x.strip()))\
                .map(lambda x: (x[0], x[1].lower()))\
                .map(lambda x: (x[0], re.sub(r'[^\w\s]','', x[1])))\
                .map(lambda x: (x[0], unicodedata.normalize('NFC',x[1])))\
                .map(lambda x: (x[0], x[1].split(' ')))\
                .map(lambda x: (x[0], list(filter(None, x[1])))).cache()
```

```python
print(names.take(10))
```

```python
tokens = names.map(lambda x: x[1]).collect()
```

```python
shingles = []
for each in tokens:
    shingles.extend(each)
shingles = sorted(list(set(shingles)))
rep_shingles = {}
i = 1
for each in shingles:
    rep_shingles[each] = i
    i = i+1
```

```python
print(rep_shingles)
```

```python
trans_names = names.mapValues(lambda x: [rep_shingles[i] for i in x]).cache()
```

```python
print(trans_names.take(20))
```

```python
def min_hash(values):
    sign = []
    for i in range(1, 31):
        hash_list = []
        for v in values:
            h_value = (2*v + 11*i) % 100
            hash_list.append(h_value)
        sign.append(min(hash_list))
    return sign
def band(item):
    band = []
    for i in range(0, 6):
        band.append((i, (item[0], item[1][i])))
    return band

# def candidate(input):
#     result = []
#     permutations = list(combinations(input, 2))
#     for each in permutations:
#         u1 = each[0]
#         u2 = each[1]
#         if operator.eq(u1[1],u2[1]):
#             candi = tuple([u1[0], u2[0]])
#             reverse_candi = tuple([u2[0], u1[0]])
#             result.append(candi)
#     return result
```

```python
signiture = trans_names.mapValues(lambda x: min_hash(x))\
                        .mapValues(lambda x: [x[i:i+5] for i in range(0, 30, 5)])\
                        .flatMap(lambda x: band(x))\
                        .groupByKey()\
                        .map(lambda x:(x[0], list(x[1]))).cache()
```

```python
print(signiture.take(2))
```

```python
def candidates(values):
    cluster = {}
    for each in values:
        if cluster.has_key(each[1]):
            cluster[each[1]].append(each[0])
        else:
            cluster[each[1]]=[]
            cluster[each[1]].append(each[0])
    return cluster
    
```

```python
candidates = signiture.mapValues(lambda values: candidates(values))
```

```python
print(candidates.take(2))
```
